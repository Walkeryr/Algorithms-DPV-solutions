Functions:

f(n) = n^1.01

g(n) = n(log(n))^2

Solution: f = O(g)

```python
from math import log
```

```python
def f(n):
    return n ** 1.01
```

```python
def g(n):
    return n * (log(n) ** 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 1400000, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
# plt.plot(n, [20 * g(i) for i in n], label='20 * g(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
