$$ f(n) = n - 100 $$

$$ g(n) = n - 200 $$

$$ Solution: f = Î˜(g) $$

```python
def f(n):
    return n - 100;
```

```python
def g(n):
    return n - 200;
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 1000, 1)

plt.xkcd()
plt.plot(n, [f(n) for n in x], label='f(n)')
plt.plot(n, [g(n) for n in x], label='g(n)')
plt.plot(n, [2 * g(n) for n in x], label='2 * g(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
