Functions:

f(n) = n^(1/2)

g(n) = n^(2/3)

Solution: f = O(g)

```python
def f(n):
    return n ** 1/2;
```

```python
def g(n):
    return n ** 2/3;
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 400, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
plt.plot(n, [100 * f(i) for i in n], label='100 * f(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
