Functions:

f(n) = n^2 / log(n)

g(n) = n(log(n))^2

Solution: f = Î©(g)

```python
from math import log
```

```python
def f(n):
    return n ** 2 / log(n)
```

```python
def g(n):
    return n * (log(n) ** 2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(2, 1400000, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
plt.plot(n, [400 * g(i) for i in n], label='400 * g(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
