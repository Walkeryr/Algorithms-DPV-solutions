Functions:

f(n) = nlog(n)

g(n) = 10nlog(10n)

Solution: f = Î˜(g)

```python
from math import log
```

```python
def f(n):
    return n * log(n)
```

```python
def g(n):
    return 10 * n * log(10 * n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 400, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
plt.plot(n, [20 * f(i) for i in n], label='20 * f(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
