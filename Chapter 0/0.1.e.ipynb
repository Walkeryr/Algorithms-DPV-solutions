Functions:

f(n) = log(2n)

g(n) = log(3n)

Solution: f = Î˜(g)

```python
from math import log
```

```python
def f(n):
    return log(2 * n)
```

```python
def g(n):
    return log(3 * n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 400, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
plt.plot(n, [2 * f(i) for i in n], label='2 * f(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
