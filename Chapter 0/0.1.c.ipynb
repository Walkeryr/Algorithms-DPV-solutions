Functions:

$$ f(n) = 100n + log(n) $$

$$ g(n) = n + (log(n))^2 $$

Solution: f = Î˜(g)

```python
from math import log
```

```python
def f(n):
    return 100 * n + log(n)
```

```python
def g(n):
    return n + log(n) ** 2
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(1, 400, 1)

plt.xkcd()
plt.plot(n, [f(i) for i in n], label='f(n)')
plt.plot(n, [g(i) for i in n], label='g(n)')
plt.plot(n, [100 * g(i) for i in n], label='100 * g(n)')
plt.ylabel('f(n)')
plt.xlabel('n')
plt.legend(loc='upper left')
plt.show()
```
